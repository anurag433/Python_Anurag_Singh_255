```python
sentence = input("Enter a sentence: ")

digits = 0
uppercase = 0
lowercase = 0

for ch in sentence:
    if ch.isdigit():
        digits += 1
    elif ch.isupper():
        uppercase += 1
    elif ch.islower():
        lowercase += 1

print("sentence:", sentence )
print("Digits:", digits)
print("Uppercase Letters:", uppercase)
print("Lowercase Letters:", lowercase)
```

```python
def hcf_two(a, b):
    while b != 0:
        a, b = b, a % b
    return a


def find_hcf(numbers):
    hcf = numbers[0]
    for n in numbers[1:]:
        hcf = hcf_two(hcf, n)   
    return hcf

nums = [24, 60, 36, 48]
print("HCF:", find_hcf(nums))
```

```python
def makePairs(list1, list2):
    
    if len(list1) != len(list2):
        return []

    
    result = []
    for i in range(len(list1)):
        result.append((list1[i], list2[i]))

    return result

print(makePairs([1, 3, 5, 7], [2, 4, 6, 8]))

```

```python
def is_perfect(num):
    sum_factors = 0
    for i in range(1, num):      # proper divisors
        if num % i == 0:
            sum_factors += i
    return sum_factors == num


def perfect_numbers_upto_1000():
    print("Perfect numbers from 1 to 1000 are:")
    for n in range(1, 1001):
        if is_perfect(n):
            print(n)


# Call the function
perfect_numbers_upto_1000()
```

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True


def print_primes_upto_N(N):
    print(f"Prime numbers between 1 and {N}:")
    for n in range(1, N + 1):
        if is_prime(n):
            print(n)


# Taking input
N = int(input("Enter a number N: "))

# Function call
print_primes_upto_N(N)
```
